```python
import numpy as np 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data =pd.read_csv('creditcard.csv')
data.head()
```

Get information of dataset

```python
data.info()
```

```python
data.isnull().sum()
```

Data staticis information get

```python
data.describe()
```

Access the class column value and 0 for legit and 1 for fraud transaction 

```python
data['Class'].value_counts()
```

```python
legit = data[data.Class==0]
fraud = data[data.Class==1]
```

define the mean of amount of legit and fraud

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
data.groupby('Class').mean()
```

```python
print(legit.shape)
print(fraud.shape)
```

data is unblaced in legit and fraud transaction then we apply sampling in the legit dataset

```python
legit_sample = legit.sample(n=492)
```

```python
print(legit_sample.shape)
```

Concating  to datasets 

```python
new_data = pd.concat([legit_sample , fraud], axis = 0)
new_data.head()
```

```python
new_data.tail()
```

```python
new_data['Class'].value_counts()
```

```python
new_data.groupby('Class').mean()
```

```python
x =new_data.drop(columns='Class', axis=1)
y =new_data['Class']
```

Train and Test part split the model

```python
X_train, X_test, y_train , y_test = train_test_split(x, y, test_size=0.3 , random_state=2)
```

```python
model =LogisticRegression()
model.fit(X_train, y_train)
```

```python
model_train_predict = model.predict(X_train)
model_train_acc = accuracy_score(model_train_predict,y_train)
```

```python
print("Accuracy score for Training the model ", model_train_acc)
```

```python
model_test_predict = model.predict(X_test)
model_test_acc = accuracy_score(model_test_predict, y_test)
```

```python
print("accuracy_score for testing model prediction", model_test_acc)
```
